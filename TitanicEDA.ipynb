## 1- IMPORT LIBRARIES

```python
# Import Libraries for EDA and DAta Analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style

sns.set(style="whitegrid")
```

## 2- LOAD THE DATASET

```python
# Load the Titanic dataset
df = pd.read_csv(r'C:\Users\Sanyam\Desktop\Sanyam\VS code\Libraries Python\titanic.csv')
df
```

```python
# Display the first few rows
df.head()
```

## 3- DATA OVERVIEW

```python
# Display basic information about the dataset
df.info()

# Show statistical summary
df.describe()

# Check for missing values
df.isnull().sum()
```

## 4- DATA CLEANING

```python
# Dropping columns that won't be used in analysis
df = df.drop(columns=['Ticket', 'Cabin', 'Name'])

# Fill missing values in 'Age' with the median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Drop rows with missing 'Embarked'
df = df.dropna(subset=['Embarked'])
```

## 5- VISUALIZE SURVIVAL RATES

## 5.1 - OVERALL SURVIVAL RATES

```python
# Overall survival rate
plt.figure(figsize=(5,5))
sns.countplot(x='Survived', data=df,color='GREEN')
plt.title('Overall Survival Rate')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.show()
```

## 5.2 - SURVIVAL RATE BY GENDER

```python
# Survival rate by gender
plt.figure(figsize=(5,5))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.show()
```

## 5.3 - SURVIVAL RATES BY PASSENGER CLASS

```python
# Survival rate by passenger class
plt.figure(figsize=(5,5))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

## 5.4 - AGE DISTRIBUTION OF SURVIVORS VS NON-SURVIVOR

```python
# Age distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', bins=30, kde=True)
plt.title('Age Distribution of Survivors vs. Non-Survivors')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## 6- CORRELATION ANALYSIS

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))

# Convert categorical variables to numeric using one-hot encoding
df = pd.get_dummies(df, drop_first=True)  

# Calculate the correlation matrix
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
